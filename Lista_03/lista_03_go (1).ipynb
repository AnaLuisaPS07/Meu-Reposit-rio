```python
**QUESTÃO 1**
package main

import "fmt"

func main() {
    var N int
    fmt.Scanln(&N)

    V := make([]int, N)
    for i := 0; i < N; i++ {
        fmt.Scan(&V[i])
    }

    var M int
    fmt.Scanln(&M)

    buscas := make([]int, M)
    for i := 0; i < M; i++ {
        fmt.Scanln(&buscas[i])
    }

    resultados := make([]string, M)
    for i, busca := range buscas {
        encontrado := false
        for _, num := range V {
            if num == busca {
                encontrado = true
                break
            }
        }
        if encontrado {
            resultados[i] = "ACHEI"
        } else {
            resultados[i] = "NAO ACHEI"
        }
    }

    for _, resultado := range resultados {
        fmt.Println(resultado)
    }
}
```

```python
**QUESTÃO 5**

package main

import "fmt"

func main() {
    var casos [][]int

    for {
        var N int
        fmt.Println("Digite o tamanho do vetor:")
        fmt.Scanln(&N)

        if N == 0 {
            break
        }

        vetor := make([]int, N)

        fmt.Println("Digite os elementos do vetor separados por espaço:")
        for i := 0; i < N; i++ {
            fmt.Scan(&vetor[i])
        }

        casos = append(casos, vetor)
    }

    fmt.Println("Saída:")
    for _, vetor := range casos {
        maior := vetor[0]
        indiceMaior := 0

        for i := 1; i < len(vetor); i++ {
            if vetor[i] > maior {
                maior = vetor[i]
                indiceMaior = i
            }
        }

        fmt.Printf("%d %d\n", indiceMaior, maior)
    }
}
```

```python
**QUESTÃO 7**

package main

import (
	"fmt"
)

func main() {
	var N int
	for {
		fmt.Scanln(&N)
		if N == 0 {
			break
		}

		var vec []int
		for i := 0; i < N; i++ {
			var num int
			fmt.Scan(&num)
			vec = append(vec, num)
		}

		max := 0
		for _, v := range vec {
			if v > max {
				max = v
			}
		}

		freq := make(map[int]int)
		for _, v := range vec {
			for i := 0; i <= max; i++ {
				if v <= i {
					freq[i]++
				}
			}
		}

		for i := 0; i <= max; i++ {
			fmt.Printf("(%d) %d\n", i, freq[i])
		}
	}
}
```

```python
**QUESTÃO 8**

package main

import (
	"fmt"
)

func main() {
	var n int

	for {

		_, err := fmt.Scan(&n)
		if err != nil {
			break
		}

		fmt.Println(decimalParaBinario(n))
	}
}

func decimalParaBinario(num int) string {
	if num == 0 {
		return "0"
	}

	bin := ""

	for num > 0 {
		resto := num % 2
		num = num / 2
		bin = fmt.Sprintf("%d%s", resto, bin)
	}

	return bin
}
```

```python
**QUESTÃO 9**

package main

import (
	"fmt"
	"math"
)

func main() {
	var N int
	fmt.Scanln(&N)

	pontos := make([][3]float64, N)
	for i := 0; i < N; i++ {
		fmt.Scanln(&pontos[i][0], &pontos[i][1], &pontos[i][2])
	}

	for i := 1; i < N; i++ {
		distancia := calcularDistancia(pontos[i-1], pontos[i])
		fmt.Printf("%.2f\n", distancia)
	}
}

func calcularDistancia(p1, p2 [3]float64) float64 {
	dx := p1[0] - p2[0]
	dy := p1[1] - p2[1]
	dz := p1[2] - p2[2]
	return math.Sqrt(dx*dx + dy*dy + dz*dz)
}
```

```python
**QUESTÃO 11**

package main

import (
	"fmt"
)

func main() {
	var N int
	fmt.Scanln(&N)

	V := make([]int, N)
	W := make([]int, N)

	for i := 0; i < N; i++ {
		fmt.Scan(&V[i])
	}

	for i := 0; i < N; i++ {
		W[i] = V[N-i-1]
	}

	maiorV, menorW := V[0], W[0]
	for _, valor := range V {
		if valor > maiorV {
			maiorV = valor
		}
	}
	for _, valor := range W {
		if valor < menorW {
			menorW = valor
		}
	}

	for i := 0; i < N; i++ {
		fmt.Printf("%d", V[i])
		if i < N-1 {
			fmt.Print(" ")
		}
	}
	fmt.Println()

	for i := 0; i < N; i++ {
		fmt.Printf("%d", W[i])
		if i < N-1 {
			fmt.Print(" ")
		}
	}
	fmt.Println()

	fmt.Println(maiorV)
	fmt.Println(menorW)
}
```

```python
**QUESTÃO 13**

package main

import (
	"fmt"
)

func main() {
	var (
		N int
		Frequencia map[int]int
		MaiorFrequencia int
		MaisFrequente int
		num int
	)

	fmt.Scanln(&N)
	Frequencia = make(map[int]int)

	for i := 0; i < N; i++ {
		fmt.Scan(&num)
		Frequencia[num]++
	}

	for num, freq := range Frequencia {
		if freq > MaiorFrequencia || (freq == MaiorFrequencia && num < MaisFrequente) {
			MaiorFrequencia = freq
			MaisFrequente = num
		}
	}

	fmt.Println(MaisFrequente)
	fmt.Println(MaiorFrequencia)
}
```

```python
**QUESTÃO 15**

package main

import (
	"fmt"
	"math"
)

func main() {
	var T int
	fmt.Scanln(&T)

	testCases := make([][]int, T)

	for t := 0; t < T; t++ {
		var N int
		fmt.Scanln(&N)

		vetor := make([]int, N)
		for i := 0; i < N; i++ {
			fmt.Scan(&vetor[i])
		}

		testCases[t] = vetor
	}

	for _, vetor := range testCases {
		distancia, comparacoes := calcularDistancia(vetor)
		fmt.Println(distancia, comparacoes)
	}
}

func calcularDistancia(vetor []int) (int, int) {
	var minDistancia = math.MaxInt32
	var comparacoes = 0

	for i := 0; i < len(vetor)-1; i++ {
		for j := i + 1; j < len(vetor); j++ {
			distanciaAtual := int(math.Abs(float64(vetor[i] - vetor[j])))
			comparacoes++
			if distanciaAtual < minDistancia {
				minDistancia = distanciaAtual
			}
		}
	}

	return minDistancia, comparacoes
}
```

```python
**QUESTÃO 18**

package main

import (
	"fmt"
)

func main() {
	var T int
	fmt.Scanln(&T)

	for t := 0; t < T; t++ {
		var cpf [11]int
		for i := 0; i < 11; i++ {
			fmt.Scan(&cpf[i])
		}

		valid := validarCPF(cpf)
		if valid {
			fmt.Println("CPF valido")
		} else {
			fmt.Println("CPF invalido")
		}
	}
}

func validarCPF(cpf [11]int) bool {
	sum1 := 0
	sum2 := 0
	for i := 0; i < 9; i++ {
		sum1 += cpf[i] * (i + 1)
		sum2 += cpf[i] * (9 - i)
	}
	digit1 := (sum1 * 10) % 11
	if digit1 == 10 {
		digit1 = 0
	}

	digit2 := (sum2 * 10) % 11
	if digit2 == 10 {
		digit2 = 0
	}

	return digit1 == cpf[9] && digit2 == cpf[10]
}
```

```python
**QUESTÃO 20**

package main

import (
	"fmt"
	"math"
)

func main() {
	var N int
	fmt.Scanln(&N)

	var maxAbsCoords []float64

	var x1, y1, z1 float64

	for i := 0; i < N; i++ {
		var x2, y2, z2 float64
		fmt.Scanln(&x2, &y2, &z2)

		if i > 0 {
			dx := x2 - x1
			dy := y2 - y1
			dz := z2 - z1
			absX := math.Abs(dx)
			absY := math.Abs(dy)
			absZ := math.Abs(dz)
			maxAbs := math.Max(math.Max(absX, absY), absZ)

			maxAbsCoords = append(maxAbsCoords, maxAbs)
		}

		x1, y1, z1 = x2, y2, z2
	}

	for _, maxAbsCoord := range maxAbsCoords {
		fmt.Printf("%.2f\n", maxAbsCoord)
	}
}
```

```python
**QUESTÃO 21**

package main

import (
	"fmt"
	"sort"
)

func main() {
	var N int
	fmt.Scanln(&N)

	pares := make([]int, 0)
	impares := make([]int, 0)

	for i := 0; i < N; i++ {
		var numero int
		fmt.Scanln(&numero)
		if numero%2 == 0 {
			pares = append(pares, numero)
		} else {
			impares = append(impares, numero)
		}
	}

	sort.Ints(pares)
	sort.Sort(sort.Reverse(sort.IntSlice(impares)))

	for _, num := range pares {
		fmt.Println(num)
	}
	for _, num := range impares {
		fmt.Println(num)
	}
}
```
